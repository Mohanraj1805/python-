```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
days = np.arange(1, 31)
health_metric = np.random.normal(loc=70, scale=10, size=30)  # e.g., blood pressure readings

# Line plot
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(days, health_metric, marker='o')
plt.title('Line Plot')
plt.xlabel('Days')
plt.ylabel('Health Metric')

# Scatter plot
plt.subplot(1, 3, 2)
plt.scatter(days, health_metric)
plt.title('Scatter Plot')
plt.xlabel('Days')
plt.ylabel('Health Metric')

# Histogram
plt.subplot(1, 3, 3)
plt.hist(health_metric, bins=10, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Health Metric')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
from google.colab import files
#uploaded = files.upload('/content/mohan.csv')
import pandas as pd

# Replace 'heart_disease.csv' with the actual file name
df = pd.read_csv('/content/mohan.csv')

# Display the first few rows of the dataframe
df.head()
import matplotlib.pyplot as plt

# Sample column names
age = df['age']
Chol = df['chol']

# Create subplots
plt.figure(figsize=(18, 5))

# Line plot
plt.subplot(1, 3, 1)
plt.plot(age, Chol, marker='o', linestyle='-')
plt.title('Line Plot of Age vs Chol')
plt.xlabel('Age')
plt.ylabel('Chol')

# Scatter plot
plt.subplot(1, 3, 2)
plt.scatter(age, Chol)
plt.title('Scatter Plot of Age vs chol')
plt.xlabel('Age')
plt.ylabel('Chol')

# Histogram
plt.subplot(1, 3, 3)
plt.hist(Chol, bins=10, edgecolor='black')
plt.title('Histogram of Chol Levels')
plt.xlabel('Chol')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
